### Encoding
Como se puede ver en los datos anteriores, los datos de correlación son solo aplicables a columnas en las cuales se tiene datos numéricos. Una solución común para esto es hacer columnas en donde se encuentren cada caso presente en esa columna. En caso de solo haber dos opciones se puede simplemente hacer una clasificación binaria (1 y 0). A este paso se le llama _encoding_.

```python
# Ver casos de la columna Embarked
titanic_2.Embarked.unique()
```

```python
titanic_2 = pd.get_dummies(titanic_2, columns=['Embarked'])
```

```python
titanic_2.head()
```

```python
# Ver casos de la columna Sex
titanic_2.Sex.unique()
```

```python
titanic_2['Sex_cat'] = titanic_2.Sex.replace(to_replace=titanic_2.Sex.unique(), value=range(2))
titanic_2.head()
```

```python
# Analizar nuevamente la importancia de las características
titanic_2.corr()['Survived']
```

### PCA
PCA es un algoritmo de aprendizaje no supervisado que permite encontrar una transformación lineal de los datos a un espacio en el cual las características resultantes estén ordenadas de menor a mayor covarianza. 
Los pasos del algoritmo son los siguientes:
* Centrado mediante resta del vector de la media $\mu$
* Cálculo de la matriz de covarianza $\Sigma$
* Obtención de los eigenvectores y eigenvalores $\lambda_i$ asociados de la matriz de covarianza.
* Ordenar los eigenvectores de acuerdo al orden de los eigenvalores, colocando de mayor a menor en la matriz $Z$.
* Obtener los nuevos valores multiplicando la matriz original por la matriz $Z$. $PC = Z^TX$

```python
from sklearn.decomposition import PCA
pc = PCA()
X_pca = pc.fit_transform(iris.as_matrix()[:,:4])
iris_pca = pd.DataFrame(data=X_pca, 
                        columns=['PC_1', 'PC_2', 'PC_3', 'PC_4'])

#Anadir la columna de especies para etiquetar
iris_pca['species'] = iris.species
```

```python
sns.pairplot(iris_pca,
            hue='species')
```

Para visualizar la importancia de cada componente podemos mostrar los eigenvalores. El método incluído en scikit-learn nos permite además obtener los normalizados, que nos dicen el porcentaje de información que posee cada uno.

```python
sns.barplot(x=['PC_1', 'PC_2', 'PC_3', 'PC_4'],
            y=pc.explained_variance_ratio_)
plt.plot()
```
